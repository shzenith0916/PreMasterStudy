### Opening the file

```python
def open_file(filename):
    with open(filename, 'r') as is_file:
        raw = [line.strip().strip('% ') for line in is_file.readlines()]
        
    return raw

credit_data = open_file('credit-g.arff')
credit_data
```

### Sample question 1: Function `attribute_names`
* Write a function that returns the names of all the attributes in the credit-g.arff file.

```python
[line for line in credit_data if line.startswith('@attribute')]
```

```python
def attribute_names():
    return [line.split(' ')[1] for line in credit_data if line.startswith('@attribute')]

attribute_names()
```

* How many observations are in each class? Write a function that returns the class distribution (counts per class).

```python
data_dict = []
for line in splitted_data:
    data_dict.append( {header:data for header,data in zip(attribute_names(), line) } )

data_dict
```

```python
def dict_data():
    data = credit_data[301:]
    splitted_data = [line.split(',') for line in data]
    
    data_dict = []
    for line in splitted_data:
        data_dict.append( {header:data for header,data in zip(attribute_names(), line) } )

    return data_dict

dict_data()
```

```python
class_data = [info.get('class','') for info in dict_data()]
class_data
```

```python
class_distribution = {class_type: class_data.count(class_type) for class_type in class_data}
class_distribution
```

```python
def class_distribution():
    class_data = [data.get('class','') for data in dict_data()]
    class_distribution = {class_type: class_data.count(class_type) for class_type in class_data}
    return class_distribution

class_distribution()
```

```python
def class_distribution2():
    class_data = [info['class'] for info in dict_data()]
    good_class = len([line for line in class_data if line == 'good'])
    bad_class = len([line for line in class_data if line == 'bad'])
    return 'good class: {}, bad class: {}'.format(good_class, bad_class)

class_distribution2()
```

### Sample question 3: Output files
* Write a function that generates an output file for each employment category. <br>
 In each file, write out the average of the credit amounts for all good loans for that employment category, rounded to two decimal places.
* Use employment category name plus .txt as the names of the output files.

```python
def employment_category():
    return set([data.get('employment','') for data in dict_data()])

employment_category()
```

```python
good_loan_credit = [float(data.get('credit_amount')) for data in dict_data() if data['class'] == 'good']
good_loan_credit
```

```python
def good_loan_credit_by_category():
    return {category : [float( data.get('credit_amount') ) for data in dict_data() 
                            if data['class'] == 'good' and data['employment'] == category ]
            for category in employment_category()}

good_loan_credit_by_category()
```

```python
def round_avg(numbers):
    return round(sum(numbers) / len(numbers),2)

def avg_good_loan_credit_by_category():
    
    return { category : round_avg( good_loan_credit_by_category()[category] ) for category in good_loan_credit_by_category()}

avg_good_loan_credit_by_category()
```

```python
def write_file_category():
    
    for category in employment_category():
        with open('{}'.format(category) + '.txt'  ,  'w') as a_file:
            
            avg_of_category = avg_good_loan_credit_by_category()[category]
            
            a_file.write(str( avg_of_category))
            
            
write_file_category()
```
