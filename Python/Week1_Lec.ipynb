## Programs
* Program
  * Sequence of instructions
  * Specifies how to perform a 'computation'
  * Typical components:
    * Input
    * Output
    * Calculation
    * Conditional execution
    * Repetition
  * Breaking a large, complex task into smaller and smaller subtasks

## Basic operators

* `+` for addition
* `-` for subtraction
* `*` for multiplication
* `/` for division
* `**` for exponentiation (raise to the power of)
* `^` for bitwise xor (you should remember that ^ is not for exponentiation in Python)

## Types and values

* int
* float (deepening: Decimal, Fraction)
* str

## Formal versus natural languages
* A programming language is a formal language
  * Rather strict syntax
  * **Parsing**: figuring out the structure. Also done by an interpreter or compiler. It is useful to be familiar with this term. Sooner or later, you will see this term in an error message.
  * Little to no ambiguity
  * Typically more concise than natural language
  * Punctuation can be important in Python it certainly is

## Expressions
* Any combination of zero or more values and/or zero or more variables and/or operators and/or function calls

```python
original_age = 14 # Assignment
years_passed = 10 # Assignment
print('Current age:', original_age + years_passed) # Expression
original_age = 16 # Variable's value may change over time, for instance if data is corrected
print('Current age:', original_age + years_passed) # Expression
```

```python
from datetime import datetime
today = datetime.now()
this_year, this_month = (today.year, today.month)
print(this_year)
print(this_month)
```

```python
import datetime

today = datetime.date.today()

print(today)

print(today.year)
print(today.month)
print(today.day)
```

```python
# Monday 0 Sunday 6
print(today.weekday())

# Moday 1 Sunday 7
print(today.isoweekday())
```

```python
tdelta = datetime.timedelta(days=7)

# 7 days later from today
print(today + tdelta)

# 7 days before from today
print(today - tdelta)
```

```python
# time = datetime.time(hour, min, sec, micro_sec)

time = datetime.time(12,30,45,3000)
print(time)
```

```python
dt = datetime.datetime(2018,10,1,12,31,35,100000)
print(dt)

print(dt.year)
print(dt.hour)
```

```python
hour_delta = datetime.timedelta(hours=5)
print(dt + hour_delta)
```

```python
dt_today = datetime.datetime.today()
dt_now = datetime.datetime.now()
dt_utcnow = datetime.datetime.utcnow()

print(dt_today)
print(dt_now)
print(dt_utcnow)
```
