```python
# Opening the file

def csv_open(filename):
    with open(filename, 'r') as csv_file:
        boxdata = [line.strip() for line in csv_file.readlines()]
    return boxdata

boxdata = csv_open('Fietstrommels-Rotterdam.csv')
boxdata
```

```python
header_box = [line.split('|') for line in boxdata][0]
data_box = [line.split('|') for line in boxdata][1:]

boxes = []
for line in data_box: 
    boxes.append({header:boxline for header,boxline in zip(header_box, line) })
    
boxes
```

```python
def dict_box():
    header_box = [line.split('|') for line in boxdata][0]
    data_box = [line.split('|') for line in boxdata][1:]
    
    boxes = []
    for line in data_box:
        boxes.append( { header:boxline for header,boxline in zip(header_box, line) })
    
    return boxes

dict_box()
```

```python
# Type in an expression that show the total number of veloboxes (in our dataset)

def nr_boxes():
    return len(dict_box())

nr_boxes()
```

```python
# How many rows do we have with a GSM?

def boxes_with_GSM():
    return [box for box in dict_box() if box['GSM'] != '']

len(boxes_with_GSM())
```

```python
# What areas (boroughs) have (at least one) velobox with a GSM associated?

def areas_with_GSM_boxes():
    return { box['Deelgemeente'] for box in boxes_with_GSM() }

areas_with_GSM_boxes()
```

```python
# velobox areas

velobox_areas = { box['Deelgemeente'] for box in dict_box() }
velobox_areas
```

```python
# How many boxes are there for each area?

def box_per_area():
    velobox_areas = { box['Deelgemeente'] for box in dict_box() }
    return [(area, len([box for box in dict_box() if box['Deelgemeente'] == area]) ) for area in velobox_areas ]

sorted([(nr,area) for area,nr in box_per_area()], reverse= True) 
```

```python
# Return all street names

def all_street_names():
    return [box['Straat'] for box in dict_box()]

street_names = all_street_names()
street_names
```

```python
all_repeating_streets = {( street, street_names.count(street) ) for street in street_names if street_names.count(street) > 1 }
all_repeating_streets  
# Be careful not to use fucntion name directly. 
# Because count() function is used with dot . ,it will cause an error without renaming the list.
```

```python
# Return all street names that are present more than once and the number of times that they appear in the file.

def repeating_streets():
    all_repeating_streets = { ( street, street_names.count(street) ) for street in street_names 
                                  if street_names.count(street) > 1 }
    
    import operator 
    return sorted(all_repeating_streets, key= operator.itemgetter(1), reverse = True)

repeating_streets()
```
