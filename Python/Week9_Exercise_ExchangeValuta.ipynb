```python
EXCHANGE_RATES = { 'USD':   1.000000, # US Dollar
                   'EUR':   0.910692, # Euro
                   'GBP':   0.816860, # Pound Sterling
                   'JPY': 104.827420, # Japanese Yen
                   'CHF':   0.988881, # Swiss Franc
                   'CNY':   6.774159 } # Chinese Yuan 
```

```python
def valid_valuta_codes():
    return EXCHANGE_RATES.keys()

valid_valuta_codes()
```

```python
def check_valid_valuta(three_letter_code):
    return three_letter_code in valid_valuta_codes()

#TEST
print(check_valid_valuta('EUR'))
print(check_valid_valuta('WON'))
```

```python
def convert_to_dollars(amount, old_valuta):
    return round(amount / EXCHANGE_RATES[old_valuta] , 3)

#TEST
print(convert_to_dollars(5,'EUR'))
```

```python
def ask_amount_and_valuta_to_exchange():
    
    while True:
    
        amount_from = float(input('How much would you like to exchange from?'))
        valuta_from = input('Which currency would you like to exchange from to USD?').upper()
        
        if not check_valid_valuta(valuta_from):
            print('Please provide a valid valuta code.')
            continue
            
        dollar_amount = convert_to_dollars(amount_from, valuta_from)
        
        if dollar_amount < 10: 
            print('Please provide an amount of at least 10 USD')
            continue
            
        break 
        
        
    return amount_from, valuta_from 

#TEST
ask_amount_and_valuta_to_exchange()
```

```python
def ask_target_valuta():
    
    while True:
            
            target_valuta = input('Which valuta would you like to exchange to?').upper()
            
            if not check_valid_valuta(target_valuta):
                print('Please provide a valid valuta code.')
                continue
                
            break
            
    return target_valuta


ask_target_valuta()
```

```python
def apply_transaction_fee(amount_in_dollars):
    
    one_percent_fee = amount_in_dollars * 0.01
    
    if one_percent_fee > 2.50:
        return amount_in_dollars * 0.99
    
    else: 
        return amount_in_dollars - 2.50
    

apply_transaction_fee(300)
```

```python
def convert_to_foreign_currency(amount_usd, new_valuta):
    
    converted = amount_usd * EXCHANGE_RATES[new_valuta]
    return converted
```

```python
def inform_client(amount,valuta):
    print('The amount you will be exchanged is {} {}'.format(amount,valuta))
```

```python
def exchange_valuta():
    from_amount, from_valuta = ask_amount_and_valuta_to_exchange()
    to_valuta = ask_target_valuta()
    amount_in_dollars = convert_to_dollars(from_amount, from_valuta)
    amount_in_dollars_after_fee = apply_transaction_fee(amount_in_dollars)
    amount_in_requested_currency = convert_to_foreign_currency(amount_in_dollars_after_fee, to_valuta)
    inform_client(amount_in_requested_currency, to_valuta)
```

```python
exchange_valuta()
```
